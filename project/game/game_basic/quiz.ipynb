```python
#[게임조건]
# 1. 캐릭터는 화면 가장 아래에 위치, 좌우로만 이동 가능
# 2. 적은 화면 가장 위에서 떨어짐, x좌표는 매번 랜덤으로 설정
# 3. 캐릭터가 적을 피하면 그 다음 적이 접근
# 4. 캐릭터가 적과 충돌하면 게임 종료
# 5. FPS는 30으로 고정

# [게임이미지]
# 1. 배경 : 640*480 (세로, 가로) - background.png
# 2. 캐릭터 : 70*70 - character.png
# 3. 적 : 70*70 - enemy.png

import sys #pygame 라이브러리가 호출이 안되서 임시방편으로 사용함
sys.path.append(r'c:\users\samsung\anaconda3\lib\site-packages')
import pygame
import random
pygame.init()

# 배경
screen_width = 480
screen_height = 640
screen = pygame.display.set_mode((screen_width,screen_height))
background = pygame.image.load(r'C:\python\project\game\background.png')

# 제목
pygame.display.set_caption("quiz")

# 캐릭터
character = pygame.image.load(r'C:\python\project\game\character.png')
character_size = character.get_rect().size
character_width = character_size[0]
character_height = character_size[1]
character_xpo = (screen_width/2) - (character_width/2)
character_ypo = screen_height - character_height
character_speed = 0.6

# enemy
enemy = pygame.image.load(r'C:\python\project\game\enemy.png')
enemy_size = enemy.get_rect().size
enemy_width = enemy_size[0]
enemy_height = enemy_size[1]
enemy_xpo = random.uniform(0,screen_width - enemy_width)
enemy_ypo = 0
enemy_speed = 10

# 이동 좌표
to_x = 0

# fps
clock = pygame.time.Clock()

# time
total_time = 10000
start_ticks = pygame.time.get_ticks()

# 이벤트
running = True
while running:
    dt = clock.tick(30)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            
            
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_RIGHT:
                to_x += character_speed
            elif event.key == pygame.K_LEFT:
                to_x -= character_speed
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_RIGHT or event.key == pygame.K_LEFT:
                to_x = 0
                
    character_xpo += to_x * dt
    enemy_ypo += enemy_speed 
    
    # 경계값 설정
    if character_xpo < 0:
        character_xpo = 0
    elif character_xpo > screen_width - character_width:
        character_xpo = screen_width - character_width
    
    # enemy 위치 초기화
    if enemy_ypo > screen_height:
        enemy_ypo = 0
        enemy_xpo = random.uniform(0, screen_width - enemy_width)
    
    # 충돌 처리 (움직일 때마다 rect 정보를 업데이트)
    character_rect = character.get_rect()
    character_rect.left = character_xpo
    character_rect.top = character_ypo
    
    enemy_rect = enemy.get_rect()
    enemy_rect.left = enemy_xpo
    enemy_rect.top = enemy_ypo 
    
    if character_rect.colliderect(enemy_rect):
        print("게임 끝")
        running = False
    
    # 구현
    screen.blit(background,(0,0))
    screen.blit(character,(character_xpo,character_ypo))
    screen.blit(enemy,(enemy_xpo,enemy_ypo))
    
    pygame.display.update()
    
# 게임 종료
pygame.time.delay(2000)
pygame.quit()
```
